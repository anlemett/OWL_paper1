```python
filename = os.path.join(DATA_DIR, "UO1_UO_neurometric_24_11_2023.csv")
df_low = pd.read_csv(filename, sep=';')

filename = os.path.join(DATA_DIR, "UO3_UO_neurometric_24_11_2023.csv")
df_high = pd.read_csv(filename, sep=';')

filename = os.path.join(DATA_DIR, "UO2_UO_neurometric_24_11_2023.csv")
df_medium = pd.read_csv(filename, sep=';')
```

```python
def getTimestampSecond(timestamp):

    return math.floor(timestamp/1000)

def getSecond(timestamp, first_timestamp):
    return timestamp - first_timestamp

df_low['timestamp'] = df_low.apply(lambda row: getTimestampSecond(row['calculatedAt']), axis=1)
first_timestamp = df_low.iloc[0, df_low.columns.get_loc('timestamp')]
df_low['second'] = df_low.apply(lambda row: getSecond(row['timestamp'], first_timestamp), axis=1)

df_high['timestamp'] = df_high.apply(lambda row: getTimestampSecond(row['calculatedAt']), axis=1)
first_timestamp = df_high.iloc[0, df_high.columns.get_loc('timestamp')]
df_high['second'] = df_high.apply(lambda row: getSecond(row['timestamp'], first_timestamp), axis=1)

df_medium['timestamp'] = df_medium.apply(lambda row: getTimestampSecond(row['calculatedAt']), axis=1)
first_timestamp = df_medium.iloc[0, df_medium.columns.get_loc('timestamp')]
df_medium['second'] = df_medium.apply(lambda row: getSecond(row['timestamp'], first_timestamp), axis=1)
```
