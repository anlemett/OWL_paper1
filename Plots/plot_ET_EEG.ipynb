```python
#def two_scales(ax1, index1, index2, data1, data2, y_label1, y_label2, data_min, data_max):
def two_scales(ax1, index1, index2, data1, data2, y_label1, y_label2):
    ax2 = ax1.twinx()
    
    lns1 = ax1.plot(index1, data1, color='b', linestyle = '--')
    lns2 = ax2.plot(index2, data2, color='r')
        
    lns = lns1 +  lns2
    ax2.legend(lns, [y_label1,y_label2], loc='best')
    
    ax1.set_xticklabels([])
        
    #ax1.set_ylim(data_min, data_max)
    ax1.set_ylabel(y_label1)
    
    #ax2.set_ylim(data_min, data_max)
    ax2.set_ylabel(y_label2)        

    for t in ax1.get_yticklabels():
        t.set_color('b')
        
    for t in ax2.get_yticklabels():
        t.set_color('r')
    
    return ax1, ax2
```

```python
def plot_eegmetric_pupdiam(eeg_df, et_df, metric, fig_name, title=""):
    plt.rcParams["figure.figsize"] = (15,6)
    plt.rcParams.update({'font.size': 22})

    ax = plt.subplot(1, 1, 1)
    ax1, ax2 = two_scales(ax, eeg_df.index, et_df.index, eeg_df[metric], et_df['av_pup_diameter'], metric, 'av.pup.diam')

    plt.title(title)
    plt.tight_layout()
    
    plt.savefig(os.path.join(FIGURES_DIR, fig_name + '.png'))
```

```python
filename = os.path.join(EEG_DATA_DIR, "UO_eeg_low_sec.csv")
eeg_df = pd.read_csv(filename, sep=' ')

filename = os.path.join(ET_DATA_DIR, "UO_et_low_sec.csv")
et_df = pd.read_csv(filename, sep=' ')

print(len(eeg_df))
print(len(et_df))
```

```python
plot_eegmetric_pupdiam(eeg_df, et_df, "av_workload", "Workload_PupDiam_low_UO", "UO: Low, Workload & Pup.Diam")
```

```python
plot_eegmetric_pupdiam(eeg_df, et_df, "av_stress", "Stress_PupDiam_low_UO", "UO: Low, Stress & Pup.Diam")
```

```python
plot_eegmetric_pupdiam(eeg_df, et_df, "av_vigilance", "Vigilance_PupDiam_low_UO", "UO: Low, Vigilance & Pup.Diam")
```
