```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import os
```

```python
DATA_DIR = os.path.join("..", "..")
DATA_DIR = os.path.join(DATA_DIR, "Data")
EEG_DIR = os.path.join(DATA_DIR, "EEG_Fetched_Data")
CH_DIR = os.path.join(DATA_DIR, "CH")
FIGURES_DIR = os.path.join("Figures", "EEG_CH")
```

```python
def two_scales(ax1, index1, index2, data1, data2, y_label1, y_label2):
    ax2 = ax1.twinx()
    
    lns1 = ax1.plot(index1, data1, color='b', linestyle = '--')
    lns2 = ax2.plot(index2, data2, color='r')
        
    lns = lns1 +  lns2
    ax2.legend(lns, [y_label1,y_label2], loc='best')
    
    ax1.set_xticklabels([])
        
    ax1.set_ylabel(y_label1)
    
    ax2.set_ylabel(y_label2)
    ax2.yaxis.set_major_locator(MaxNLocator(integer=True))

    for t in ax1.get_yticklabels():
        t.set_color('b')
        
    for t in ax2.get_yticklabels():
        t.set_color('r')
    
    return ax1, ax2
```

```python
def plot_eeg_ch(df_high_eeg, df_high_ch, metric, fig_name, title=""):
    plt.rcParams["figure.figsize"] = (15,6)
    plt.rcParams.update({'font.size': 22})

    ax = plt.subplot(1, 1, 1)
    ax1, ax2 = two_scales(ax, df_high_eeg["calculatedAt"], df_high_ch["timestamp"], df_high_eeg[metric], df_high_ch["score"], metric, 'CH')

    plt.title(title)
    plt.tight_layout()
    
    
    plt.savefig(os.path.join(FIGURES_DIR, fig_name + '.png'))
```

# Day 1, MO

```python
filename = os.path.join(EEG_DIR, "MO3_MO_neurometric_21_11_2023.csv")
df_eeg_high = pd.read_csv(filename, sep=';')
df_eeg_high["calculatedAt"] = df_eeg_high["calculatedAt"].apply(lambda x:int(x/1000))
df_eeg_high = df_eeg_high.drop_duplicates(subset=['calculatedAt'])

filename = os.path.join(CH_DIR, "D1r3_MO_231121_093949.csv")
df_ch_high = pd.read_csv(filename, sep=' ')
```

```python
plot_eeg_ch(df_eeg_high, df_ch_high, "stress", "Stress_CH__MO", "MO: High, Stress & CH")
```

```python
plot_eeg_ch(df_eeg_high, df_ch_high, "vigilance", "Vigilance_CH__MO", "MO: High, Vigilance & CH")
```

```python
plot_eeg_ch(df_eeg_high, df_ch_high, "workload", "Workload_CH__MO", "MO: High, Workload & CH")
```
