# Functions for Plotting

```python
def one_scale(ax1, index, data, y_label, data_min, data_max):

    lns = ax1.plot(index, data, color='r')
        
    ax1.legend(lns, [y_label], loc='best')

    ax1.set_xticklabels([])
    
    ax1.set_ylim(data_min, data_max)
    ax1.set_ylabel(y_label)
    
    for t in ax1.get_yticklabels():
        t.set_color('r')
        
    return ax1

def two_scales(ax1, index1, index2, data1, data2, y_label1, y_label2, data_min, data_max):
    ax2 = ax1.twinx()
    
    lns1 = ax1.plot(index1, data1, color='b', linestyle = '--')
    lns2 = ax2.plot(index2, data2, color='r')
        
    lns = lns1 +  lns2
    ax2.legend(lns, [y_label1,y_label2], loc='best')
    
    ax1.set_xticklabels([])
        
    ax1.set_ylim(data_min, data_max)
    ax1.set_ylabel(y_label1)
    
    ax2.set_ylim(data_min, data_max)
    ax2.set_ylabel(y_label2)        

    for t in ax1.get_yticklabels():
        t.set_color('b')
        
    for t in ax2.get_yticklabels():
        t.set_color('r')
    
    return ax1, ax2
```

```python
def plot_metric(df_low, df_high, metric, fig_name, title=""):
    plt.rcParams["figure.figsize"] = (15,6)
    plt.rcParams.update({'font.size': 22})

    data_min = min(df_high[metric].min(), df_low[metric].min())
    data_max = max(df_high[metric].max(), df_low[metric].max())

    data_max = data_max + 0.05
    data_min = data_min - 0.05

    ax = plt.subplot(1, 1, 1)
    ax1, ax2 = two_scales(ax, df_low.index, df_high.index, df_low[metric], df_high[metric], 'low', 'high', data_min, data_max)

    plt.title(title)
    plt.tight_layout()
    
    plt.savefig(os.path.join(FIGURES_DIR, fig_name + '.png'))
```

# Day 1, MO

```python
filename = os.path.join(DATA_DIR, "MO1_MO_neurometric_21_11_2023.csv")
df_low = pd.read_csv(filename, sep=';')

filename = os.path.join(DATA_DIR, "MO3_MO_neurometric_21_11_2023.csv")
df_high = pd.read_csv(filename, sep=';')
```

```python
plot_metric(df_low, df_high, "workload", "Workload_MO", "MO: High & Low, Workload")
```

```python
plot_metric(df_low, df_high, "stress", "Stress_MO", "MO: High & Low, Stress")
```

```python
plot_metric(df_low, df_high, "vigilance", "Vigilance_MO", "MO: High & Low, Vigilance")
```
